```python
import pandas as pd


# Replace 'your_file.xlsx' with the actual path to your Excel file
file_path = '12024.xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)

# Display the DataFrame (optional)
df.head()
```

```python
# create new columns, 'year', 'month', 'day' from the 'Date' column
df['Date'] = pd.to_datetime(df['Date'], format='%d %b %Y')
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day

# Display the DataFrame with the new columns (optional)
df.head()
```

```python
import matplotlib.pyplot as plt

# Create a sub plot of the polarity and subjectivity for each year and month
fig, ax = plt.subplots(figsize=(10, 6))

# Group the DataFrame by year and month, and calculate the mean polarity
df.groupby(['year', 'month'])['polarity'].mean().plot(ax=ax)

# Set the title and labels
ax.set_title('Average Polarity by Year and Month')
ax.set_xlabel('Year and Month')
ax.set_ylabel('Average Polarity')

# Display the plot
plt.show()


```

```python
import matplotlib.pyplot as plt

#create a chart that showing the average polarity by year and month in bar chart format and subjectivity in line chart format without using subplot
fig, ax = plt.subplots(figsize=(10, 6))

# Group the DataFrame by year and month, and calculate the mean polarity
df.groupby(['year', 'month'])['polarity'].mean().plot(kind='bar', ax=ax, color='blue')

# Group the DataFrame by year and month, and calculate the mean subjectivity
df.groupby(['year', 'month'])['subjectivity'].mean().plot(kind='line', ax=ax, color='red')

# Set the title and labels
ax.set_title('Average Polarity and Subjectivity by Year and Month')
ax.set_xlabel('Year and Month')
ax.set_ylabel('Average Polarity and Subjectivity')

# Display the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(30, 18))

# Group the DataFrame by year and month, and calculate the mean polarity
df.groupby(['year', 'month','day'])['polarity'].mean().plot(ax=ax)
df.groupby(['year', 'month','day'])['subjectivity'].mean().plot(ax=ax)

# Set the title and labels
ax.set_title('Average Polarity and Subjectivity by Year and Month and Day')
ax.set_xlabel('Year and Month and Day')
ax.set_ylabel('Average Polarity and Subjectivity')

# Display the plot
plt.show()
```

```python
##Display the distribution of polarity or subjectivity scores.
#creata a histogram chart that showing the distribution of polarity and subjectivity
fig, ax = plt.subplots(figsize=(10, 6))

# Create a histogram of the polarity
df['polarity'].plot(kind='hist', ax=ax, color='blue', alpha=0.7)

# Create a histogram of the subjectivity
df['subjectivity'].plot(kind='hist', ax=ax, color='red', alpha=0.7)

# Set the title and labels
ax.set_title('Distribution of Polarity and Subjectivity')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')

#add legend
ax.legend(['Polarity', 'Subjectivity'])


# Display the plot
plt.show()

```

```python
#Visualize the distribution of polarity or subjectivity scores, showing quartiles and outliers.
#create a boxplot chart that showing the distribution of polarity and subjectivity

fig, ax = plt.subplots(figsize=(10, 6))

# Create a boxplot of the polarity
df['polarity'].plot(kind='box', ax=ax, color='blue')

# Create a boxplot of the subjectivity
df['subjectivity'].plot(kind='box', ax=ax, color='red')

# Set the title and labels
ax.set_title('Distribution of Polarity and Subjectivity')
ax.set_ylabel('Value')

#add legend
ax.legend(['Polarity', 'Subjectivity'])

# Display the plot
plt.show()
```

```python
##Explore the relationship between polarity and subjectivity.

#create a scatter plot chart that showing the relationship between polarity and subjectivity
fig, ax = plt.subplots(figsize=(50, 30))

# Create a scatter plot of the polarity and subjectivity
df.plot(kind='scatter', x='polarity', y='subjectivity', ax=ax)

# Set the title and labels
ax.set_title('Polarity vs. Subjectivity')
ax.set_xlabel('Polarity')
ax.set_ylabel('Subjectivity')

#add legend
ax.legend(['Polarity', 'Subjectivity'])

# Display the plot
plt.show()
```

```python
#create a chart that shows the total count of negative polarity, neutral polarity, and positive polarity 
#the chart need to be in bar chart
#group by year and month in each sub plot

fig, ax = plt.subplots(figsize=(10, 6))

# Group the DataFrame by year and month, and calculate the count of negative polarity
df[df['polarity'] < 0].groupby(['year', 'month'])['polarity'].count().plot(kind='line', ax=ax, color='red')

# Group the DataFrame by year and month, and calculate the count of neutral polarity
df[df['polarity'] == 0].groupby(['year', 'month'])['polarity'].count().plot(kind='line', ax=ax, color='blue')

# Group the DataFrame by year and month, and calculate the count of positive polarity
df[df['polarity'] > 0].groupby(['year', 'month'])['polarity'].count().plot(kind='line', ax=ax, color='green')

# Set the title and labels
ax.set_title('Count of Negative, Neutral, and Positive Polarity by Year and Month')
ax.set_xlabel('Year and Month')

#add legend
ax.legend(['Negative', 'Neutral', 'Positive'])

# Display the plot
plt.show()

```

```python
#Represent the density of sentiment or subjectivity scores.
#create a density chart that showing the density of polarity and subjectivity
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))

# Create a density plot of the polarity
df['polarity'].plot(kind='kde', ax=ax, color='blue')

# Create a density plot of the subjectivity
df['subjectivity'].plot(kind='kde', ax=ax, color='red')

# Set the title and labels
ax.set_title('Density of Polarity and Subjectivity')
ax.set_xlabel('Value')
ax.set_ylabel('Density')

#add legend
ax.legend(['polarity', 'subjectivity'])

# Display the plot
plt.show()
```

```python
#Display multiple dimensions of sentiment (e.g., positive, negative, neutral) in a radial chart.

#create a radial chart that showing the multiple dimensions of sentiment (e.g., positive, negative, neutral)
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame with the count of negative, neutral, and positive polarity
df_polarity = pd.DataFrame({
    'count': [
        df[df['polarity'] < 0].shape[0],
        df[df['polarity'] == 0].shape[0],
        df[df['polarity'] > 0].shape[0]
    ]
}, index=['Negative', 'Neutral', 'Positive'])

# Create a radial chart
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
theta = np.linspace(0, 2 * np.pi, len(df_polarity), endpoint=False).tolist()
bars = ax.bar(theta, df_polarity['count'], color='blue', alpha=0.7)

# Set the title and labels
ax.set_title('Sentiment Distribution')
ax.set_xticks(theta)
ax.set_xticklabels(df_polarity.index)

# Display the plot
plt.show()
```

```python
#convert value from Cleansed column to string
df['Cleansed'] = df['Cleansed'].astype(str)

#reason why need to do this is when generating the word cloud, return a TypeError: expected string but getting float
```

```python
#Visualize frequently occurring words in the text data, with font size indicating frequency.

#create a word cloud chart that showing the frequently occurring words in the text data, with font size indicating frequency
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, max_words=200, background_color='white').generate(' '.join(df['Cleansed']))

# Display the WordCloud
fig, ax = plt.subplots(figsize=(10, 6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_title('Word Cloud')
ax.axis('off')

# Display the plot
plt.show()
```

```python
def devicecomparepolarity(deviceName1, deviceName2, spec):
    fig, ax = plt.subplots(figsize=(10, 6))

    df[df['PDName'] =='Samsung Galaxy S22 Ultra 5G']

    # Group the DataFrame by year and month, and calculate the count of positive polarity
    df[df['PDName'] =='Samsung Galaxy S23 Ultra']

    if spec is None:
        df[(df['PDName'] ==f'{deviceName1}') & df['Cleansed']]['polarity'].plot(kind='kde', ax=ax, color='blue')
        df[(df['PDName'] ==f'{deviceName2}') & df['Cleansed']]['polarity'].plot(kind='kde', ax=ax, color='red')

    else:
        # Create a density plot of the polarity
        df[(df['PDName'] ==f'{deviceName1}') & df['Cleansed'].str.contains(f'{spec}')]['polarity'].plot(kind='kde', ax=ax, color='blue')
        df[(df['PDName'] ==f'{deviceName2}') & df['Cleansed'].str.contains(f'{spec}')]['polarity'].plot(kind='kde', ax=ax, color='red')

    # Set the title and labels
    if spec is None:
        ax.set_title(f'Positive Reviews between 2 devices')
    else:
        ax.set_title(f'Positive Reviews between 2 devices on {spec}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Density')

    #add legend
    ax.legend([f'{deviceName1}', f'{deviceName2}'])

    # Display the plot
    if spec is None:
        plt.savefig(f'{deviceName1.replace(" ","")}_{deviceName2.replace(" ","")}.png')
    else:
        plt.savefig(f'{deviceName1.replace(" ","")}_{deviceName2.replace(" ","")}_{spec}.png')
    plt.show()
    
devicecomparepolarity('Samsung Galaxy S22 Ultra 5G', 'Samsung Galaxy S23 Ultra', 'display')
devicecomparepolarity('Samsung Galaxy S22 Ultra 5G', 'Samsung Galaxy S23 Ultra',None)
```

```python
def createworldcloud(df,deviceName):
    # Create a WordCloud object
    wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(' '.join(df))

    # Display the WordCloud
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.set_title('Word Cloud')
    ax.axis('off')

    # Display the plot
    plt.savefig(f'wc_{deviceName.replace(" ","")}.png')

df_temp = df[(df['PDName'] =='Samsung Galaxy S23 Ultra')]['Cleansed']
df_temp = df_temp.convert_dtypes(convert_string=True)
df_temp.dropna(inplace=True)
createworldcloud(df_temp,'Samsung Galaxy S23 Ultra')

df_temp = df[(df['PDName'] =='Samsung Galaxy S22 Ultra 5G')]['Cleansed']
df_temp = df_temp.convert_dtypes(convert_string=True)
df_temp.dropna(inplace=True)
print(df_temp.info())
createworldcloud(df_temp, 'Samsung Galaxy S22 Ultra 5G')
```
